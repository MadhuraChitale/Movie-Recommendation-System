```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
import ast
#ast.literal_eval will convert string to a list
#for exmaple
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# text contains dictionaries
# each  string in text is converted to a list using ast.literal_eval
def convert_to_list(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
# dropping null values
movies.dropna(inplace=True)
```

```python
#convert  genres of each movie to a list
movies['genres'] = movies['genres'].apply(convert_to_list)
```

```python
#convert  keywords of each movie to a list
movies['keywords'] = movies['keywords'].apply(convert_to_list)
movies.head()
```

```python
# gets the top 3 actors from top 3 dictionaries and converts into list
def convert_top5_cast(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter != 5:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L 
```

```python
# call fun convert3 to get the top 3 actors for each movie
movies['cast'] = movies['cast'].apply(convert_top5_cast)

movies.head()
print(movies['cast'])
```

```python
# get the name of director in crew
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python

movies.sample(5)
```

```python
def remove_spaces(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
# concatentaion
movies['cast'] = movies['cast'].apply(remove_spaces)
movies['crew'] = movies['crew'].apply(remove_spaces)
movies['genres'] = movies['genres'].apply(remove_spaces)
movies['keywords'] = movies['keywords'].apply(remove_spaces)
```

```python
movies.head()
```

```python
# splitting the overview
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies['overview']
```

```python
# get the tags
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# new contains the final data frame that will be used.
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
```

```python
# convert tag to string
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
# gets the root/stem words 
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)


new['tags'].apply(stem)
```

```python
# used for vectorization
from sklearn.feature_extraction.text import CountVectorizer
#get 5000 most frequent words and exclude stop words (in,as,for, and, ..etc)
cv = CountVectorizer(max_features=5000,stop_words='english')

    
```

```python
vector = cv.fit_transform(new['tags']).toarray()
print(cv)
print(vector)
```

```python
# cosine theta is used to get the distance between the movie vectors, less the distance, more is the similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
# similarity contains distance of each movie with all other movies
# value of similarity ranges from 0 to 1, value closer to 1 implies more similarity
print(similarity.shape)
similarity
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
# enumerate(similarity[index])  [0,0.02],[1,0.8][3,1]
    
    # print(distances)
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        


    
```
